```python
def find_outliers_zscore(data:object, column:str) -> 'str':
    
    z_score = np.abs((data[column] - data[column].mean() / data[column].std()))
    outliers = data[(z_score<2)]
    return outliers
print('done')
```

```python
method_outliers = find_outliers_zscore(df, 'Estimated prevalence of TB (all forms) per 100 000 population')
print(type(method_outliers))
print('this is the shape: ', method_outliers.shape)
print('this is the type: ', type(method_outliers))
print('this is the top 5: ', method_outliers.head(3))
```

```python
def exclude_outliers_zscore(data:object, column:str) -> 'str':    

    z_score = np.abs((data[column] - data[column].mean() / data[column].std()))
    df_excl_outliers = data[(z_score>=2)]
    return df_excl_outliers
print('done')

df_excl_outliers = exclude_outliers_zscore(df, 'Estimated prevalence of TB (all forms) per 100 000 population')
df_excl_outliers.shape
```

```python
df = df_excl_outliers
df
```

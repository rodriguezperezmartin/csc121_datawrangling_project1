```python
df.drop('Method to derive TBHIV estimates', axis=1, inplace=True)
```

```python
def find_outliers_zscore(data:object, column:str) -> 'str':

    mean = data[column].mean()
    std_dev = data[column].std()
    
    z_score = np.abs((data[column] - mean) / std_dev)
    
    outliers = data[z_score < 2]
    
    return outliers
print('done')
```

```python
outliers = find_outliers_zscore(df,'Estimated mortality of TB cases (all forms, excluding HIV) per 100 000 population')
print(outliers)
print('Type: ', type(outliers))
print('Shape: ', outliers.shape)
```

```python
def exclude_outliers_zscore(data:object, column:str) -> 'str':
 
    mean = data[column].mean()
    std_dev = data[column].std()
    
    z_score = np.abs((data[column] - mean) / std_dev)
    
    df_excl_outliers = data[z_score >= 2]
    
    return df_excl_outliers

excl_outliers = exclude_outliers_zscore(df, 'Estimated mortality of TB cases (all forms, excluding HIV) per 100 000 population')
```

```python
df = excl_outliers
df
```

```python
sns.relplot(x="Year", y="Estimated mortality of TB cases (all forms, excluding HIV) per 100 000 population", hue="Year", size="Estimated prevalence of TB (all forms) per 100 000 population", 
           sizes=(50, 500), alpha=.5, palette='muted', height=6, data=df)
```
